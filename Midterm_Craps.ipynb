```python
import random
import time
firstBet = True
maxBet = 1000
#--------------------------------------------------------------------------------------------------------------#
class Bank:
    def __init__(self, bank):
        self.bank = bank
    def getBank(self):
        return self.bank
    def updateBank_buyIn(self, buyIn):
        self.bank = self.bank + buyIn
    def updateBank_bet(self,bet):
        self.bank = self.bank - bet 
    def updateBank_win(self, win):
        self.bank = self.bank + win 
    def updateBank_loss(self,loss):
        self.bank = self.bank - loss
#--------------------------------------------------------------------------------------------------------------#
def isFirstBet():
    if firstBet == True:
        return True
    else:
        return False
#--------------------------------------------------------------------------------------------------------------#

#--------------------------------------------------------------------------------------------------------------# 
def buyIn():
    return int(input("Please enter your buy in: "))
#--------------------------------------------------------------------------------------------------------------#
def rollDice():
    return [random.randint(1,6), random.randint(1,6)]
#Function to determine if the user input is valid. This function checks for the following situations
# 1) The user entered a bet greater than their buy in
# 2) The user entered a non-integer value
# 3) The user entered a bet greater than the maximum allowed
#--------------------------------------------------------------------------------------------------------------#
def isValidInput(bet):
    #check if the player has enough cash in bank
    if bet > bank.getBank():
        print("You do not have enough chips to place this bet. ")
        return False
    elif isinstance(bet,int) == False:
        print("Please enter a valid bet: Must be an integer value")
        return False
    elif bet > maxBet:
        print("Please enter a valid bet: Cannot be greater than maximum bet ($1000)")
        return False
    else:
        bank.updateBank_bet(bet)
        return True
#--------------------------------------------------------------------------------------------------------------#
def PassLineBet(bet):
    x = rollDice()
    print("Come out roll... " + str(x))
    sumX = x[0]+x[1]
    
    if sumX == 2 or sumX == 3 or sumX == 12:
        bank.updateBank_bet(bet)
        print("--------------------------------------")
        print("| You lose: $"+str(bet)+" | "+"Your bank is: $"+str(bank.getBank()) + " |")
        print("--------------------------------------")
    elif sumX == 7 or sumX == 11:
        bank.updateBank_win(bet)
        print("------------------------------------")
        print("| You Win: $" + str(bet) + " | " + "Your bank is: $" + str(bank.getBank()) + " |")
        print("------------------------------------")
    else:
        print("POINT IS SET TO " + str(sumX))
        setPoint(sumX,bet,True,"passBet")
#--------------------------------------------------------------------------------------------------------------#       
def DoNotPassLineBet(bet):
    x = rollDice()
    print("Come out roll... " + str(x))
    sumX = x[0]+x[1]
    
    if sumX == 7 or sumX == 11:
        bank.updateBank_bet(bet)
        print("--------------------------------------")
        print("| You lose: $"+str(bet)+" | "+"Your bank is: $"+str(bank.getBank()) + " |")
        print("--------------------------------------")
    elif sumX == 2 or sumX == 3 or sumX == 12:
        bank.updateBank_win(bet)
        print("------------------------------------")
        print("| You Win: $" + str(bet) + " | " + "Your bank is: $" + str(bank.getBank()) + " |")
        print("------------------------------------")
    else:
        print("POINT IS SET TO " + str(sumX))
        setPoint(sumX,bet,True, "doNotPassBet")
#--------------------------------------------------------------------------------------------------------------#    
def setPoint(point, bet, gameOn, betType):
    
    while gameOn:
        
        if betType == "passBet":
            x = rollDice()
            sumX = x[0]+x[1]
            print("Rolling dice..."+str(x))
            print("")
            time.sleep(1)

            if sumX == 7:
                print("Shooter rolled " + str(sumX))
                print("")
                gameOn = False
                bank.updateBank_bet(bet)
                print("-------------------------------------")
                print("| You Lose: $"+str(bet)+" | "+"Your bank is: $"+str(bank.getBank()) + " |")
                print("-------------------------------------")
                time.sleep(1)
            elif sumX == point:
                print("Shooter rolled " + str(sumX))
                bank.updateBank_win(bet)
                print("------------------------------------")
                print("| You Win: $" + str(bet) + " | " + "Your bank is: $" + str(bank.getBank()) + " |")
                print("------------------------------------")
                time.sleep(1)
                gameOn = False
            else:
                print ("Shooter rolled a " + str(sumX) + " ... Roll again")
                print("")
        elif betType == "doNotPassBet":
            x = rollDice()
            sumX = x[0]+x[1]
            print("Rolling dice..."+str(x))
            print("")
            time.sleep(1)

            if sumX == 7:
                print("Shooter rolled " + str(sumX))
                gameOn = False
                bank.updateBank_win(bet)
                print("-------------------------------------")
                print("| You Win: $"+str(bet)+" | "+"Your bank is: $"+str(bank.getBank()) + " |")
                print("-------------------------------------")
                time.sleep(1)
            elif sumX == point:
                print("Shooter rolled " + str(sumX))
                bank.updateBank_bet(bet)
                print("------------------------------------")
                print("| You Lose: $" + str(bet) + " | " + "Your bank is: $" + str(bank.getBank()) + " |")
                print("------------------------------------")
                time.sleep(1)
                gameOn = False
            else:
                print ("Shooter rolled a " + str(sumX) + " ... Roll again")
#--------------------------------------------------------------------------------------------------------------#
def printMainMenu():
    print("------------------MENU--------------------------")
    print("1) Pass Line Bet")
    print("2) Do Not Pass Line Bet")
    print("3) Increase buy in")
    print("4) Print bank")
    print("5) Cash out")
#--------------------------------------------------------------------------------------------------------------#
def startGame():    
    #bool flag to determine if the player wants to cash out
    cashOut = False
    
    #main loop 
    while cashOut == False:
    
        printMainMenu()
        choice = int(input())

        if choice == 1:
            bet = int(input("Please enter bet: "))
            #bank.getBank()
            
            #Check that the player has enough cash in bank to place bet
            if bet > bank.getBank():
                print("--------------------------------------------")
                print("| Please increase your buy in to play again |")
                print("--------------------------------------------")
                startGame() #go back to main menu
            PassLineBet(bet)
            #print("Your bank is : $" + str(bank.getBank()))
            
        elif choice == 2:

            bet = int(input("Please enter bet: "))
            
            #Check that the player has enough cash in bank to place bet
            if bet > bank.getBank():
                print("--------------------------------------------")
                print("| Please increase your buy in to play again |")
                print("--------------------------------------------")
                startGame() #go back to main menu
            
            DoNotPassLineBet(bet)
        
        elif choice == 3:
            bank.updateBank_buyIn(buyIn())
            print("-----------------------")
            print("| New bank is : $"+str(bank.getBank())+" |")
            print("-----------------------")
            
        elif choice == 4:
            print("-----------------------------")
            print("| You have $"+str(bank.getBank())+" in the bank |")
            print("-----------------------------")
            
        elif choice == 5:
            cashOut = True
            print("------------------------------------------------------")
            print("| Cashing out... total is $" + str(bank.getBank()) + " | Thanks for Playing! |")
            print("------------------------------------------------------")
        else:
            print("---------------------------------")
            print("| Please enter a valid selection |")
            print("---------------------------------")
#---------------------------------------------MAIN--------------------------------------------------------------#

#instantiate the object 'bank' for class Bank(self, bank). This sets the player's bank to their buy in
bank = Bank(buyIn())
#start the craps game
startGame()
```
